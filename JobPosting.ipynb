```python
numerical_cols = [cname for cname in pfeatures.columns if 
                pfeatures[cname].dtype in ['int64', 'float64']]

categorical_cols = [cname for cname in pfeatures.columns if 
                   pfeatures[cname].dtype == "object"]
```

```python
#prueba de la cardinalidad
#A las columnas con baja cardinalidad,se le aplicara un One-Hot-Encoder
low_cardinality_cols = [col for col in categorical_cols if pfeatures[col].nunique() < 10]

#A las columnas con alta cardinalidad, se le aplicara un label_encoder
high_cardinality_cols = list(set(categorical_cols)-set(low_cardinality_cols))

low_cardinality_cols
```

```python
# elimino departament de pfeatures y de high_cardinality_cols
# porque tiene mas del 60% de los valores como NaN
pfeatures = pfeatures.drop("department",axis=1)

high_cardinality_cols.remove("department")
print(high_cardinality_cols)


# elimino salary_range de pfeatures y de high_cardinality_cols
# porque tiene mas del 60% de los valores como NaN

pfeatures = pfeatures.drop("salary_range",axis=1)

high_cardinality_cols.remove("salary_range")
print(high_cardinality_cols)
```

```python
# para la variable "employment_type" voy a relleanr los valores vacios con un "Desconocido" poque hay demasiados valores vacios para remplazarlos con la media
pfeatures["employment_type"].fillna(value="Desconocido" , inplace=True)
```

```python
# Para la variable "required_experience" remplazo los valores desconocnidos con la meda ya que no son tantos
pfeatures["required_experience"].fillna(pfeatures["required_experience"].mode()[0], inplace = True)
```

```python
#ahora voy a remplazar los valores por "Desconocido" en el resto de los valores
for cols in high_cardinality_cols:
    pfeatures[cols].fillna(value= "Desconocido", inplace = True)
```
